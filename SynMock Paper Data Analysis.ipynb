```python
#need to measure index bleed, lets start with ion dataset
import pandas as pd
import numpy as np
from natsort import natsorted
#now load sorted "raw" OTU table into dataframe
df = pd.read_csv('ion.final.txt', sep='\t', index_col=0)
#print list(df.columns.values)
#print len(df)
#okay, so 1310 OTUs in entire data set, lets trim this down to only run 3, as that one is labeled here with biomock
keepers = ['m3-301-1', 'm3-301-2', 'm3-500-1', 'm3-500-2', 'm3-712-1', 'm3-712-2', 
           'm3-736-1', 'm3-736-2', 'm3-744-1', 'm3-744-2', 'm3-755-1', 
           'm3-757-2', 'm3-766-1', 'm3-766-2', 'm3-mock3-3200a', 'm3-mock3-16000', 'm3-mock3-32000b', 'm3-stds']
mock_samples = ['m3-mock3-3200a', 'm3-mock3-16000', 'm3-mock3-32000b', 'm3-stds']
samples = [item for item in keepers if item not in mock_samples]
#print samples
df2 = df.reindex(columns=keepers, index=natsorted(df.index))
df2.to_csv('ion3.bleed.otu_table.txt', sep='\t')

#now drop OTUs where they are all zero
df3 = df2.loc[(df2 != 0).any(1)]
#print df3
mock_otus = []
for x in df3.index:
    if x.startswith('mock'):
        if not x.startswith('mock3'): #mock3 is actually in some of the samples so can't use for index-bleed measurment
            mock_otus.append(x)
#print mock_otus
#now slice into two tables containing only mocks and calculate index-bleed
bleed_counts = df3.reindex(columns=samples, index=mock_otus)
bleed_sum = bleed_counts.values.sum()
real_counts = df3.reindex(columns=mock_samples, index=mock_otus)
real_sum = real_counts.values.sum()
index_bleed = int(bleed_sum) / float(real_sum) * 100
print "Index-bleed out of mock: %f%% (%i out of %i)" % (index_bleed, bleed_sum, real_sum)

#now try to to do the opposite, count number of times that samples move into mock
#first generate list of mock otus
sample_otus = []
for x in df3.index:
    if not x.startswith('mock'):
        sample_otus.append(x)
#slice first to samples, further trim list of only things found in mocks, would be spurious or chimeras
sampledf = df3.reindex(columns=samples, index=sample_otus)
sampledf = sampledf.loc[(sampledf != 0).any(1)]
cleaned_otus = sampledf.index
#print len(cleaned_otus), len(sample_otus)

#now using new list grab counts in mock
bleed_counts = df3.reindex(columns=mock_samples, index=cleaned_otus)
bleed_counts = bleed_counts.loc[(bleed_counts != 0).any(1)]
#print bleed_counts
bleed_sum = bleed_counts.values.sum()
#print len(bleed_counts.index), bleed_sum

#finally go back to samples and get read counts for the ovlerlapping otus
real_counts = df3.reindex(columns=samples, index=bleed_counts.index)
real_sum = real_counts.values.sum()
index_bleed2 = int(bleed_sum) / float(real_sum) * 100
print "Index-bleed into mock: %f%% (%i out of %i)" % (index_bleed2, bleed_sum, real_sum)
print "Total index-bleed: %f%%" % (index_bleed + index_bleed2)
```
